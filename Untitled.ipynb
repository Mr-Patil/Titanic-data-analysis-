```python
#As explored earlier Age is having approx 20% of null data.
#According to the correlation analysis , we found that pclass and Age are highly correlated
#So we will try to replace the null Age values with the help of Pclass values
plt.figure(figsize=(12,7))
sns.boxplot(x='Pclass',y='Age', data=df)
```

```python
#According to the above plots we found that people travelling in the 1st  pclass were approx 38 Age
#those in 2nd Class were of 28, and the 3rd class were 25
#writing a function for the same

def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 38

        elif Pclass == 2:
            return 28

        else:
            return 25

    else:
        return Age
```

```python
#replacing the null values
df['Age'] = df[['Age', 'Pclass']].apply(impute_age,axis=1)
```

```python
#checking if the data is replaced
sns.heatmap(df.isnull(),yticklabels= False, cbar= False, cmap= 'viridis')
```
